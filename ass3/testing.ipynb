## Preparing algorithm setup


```python
import itertools
from Annealer import Annealer
import pickle

from sklearn.svm import SVC
from utils import *

PATH = "../data_processed/breast-cancer-diagnostic.shuf.lrn.csv"
feature_structure_file =  open('../data_processed/breast-cancer_column_types.pkl', 'rb')
feature_structure = pickle.load(feature_structure_file)

data = pd.read_csv(PATH)

s = SVC()

param_grid1 = {'coef0': [0.0, 0.5, 1.0],
               'degree': [2, 3, 4],
               'gamma': ['scale', 'auto'],
               'kernel':['linear', 'poly', 'rbf', 'sigmoid'],
               'C': [0.1, 1, 10, 100]}
```

```python
Annealer.required_keys.issubset(feature_structure.keys())
```

```python
Annealer_SVC = Annealer(method= SVC(),
                        feature_structure=feature_structure,
                        search_space=param_grid1,
                        data=data)
Annealer_SVC.check_param_space()
```
